# Bouncing-ball RNN
## 

Est-ce que ça fait mieux qu'un MLP ?

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import TimeDistributed, Dense, Dropout, SimpleRNN, RepeatVector, Flatten
from tensorflow.keras.callbacks import EarlyStopping, LambdaCallback

import pickle

print('Tested with tensorflow version 2.0.1')
print('Using tensorflow version:', tf.__version__)
```

## 2: Charger les données


#### Charger les données. Reshape X pour avoir un séquence de 3 coordonnées
ex.: [ 51  92  73 172 193 192] => [[51,92], [73,172], [193, 192]]

```python
def LoadData():
    with open("x.pickle", "rb") as f:
        X = pickle.load(f)
        #print(X)
            
    with open("y.pickle", "rb") as f:
        y = pickle.load(f)
        #print(y)

    return X,y
```

```python
X,y = LoadData()
X = np.array(X)
y = np.array(y)
print(X[0], y[0])
print(X.shape, y.shape)
```

#### transformer X
Ajouter des [1,1] à chaque point. Sert d'hypothèse pour la vélocité qui devra être prédite

**note**: le input est [x1,y1,x2,y2,x3,y3] 
 
sera: 
*  [x1,y1,1,1]
*  [x2,y2,1,1]
*  [x3,y3,1,1]

Le output reste [x4,y4,vx,vy]

```python
X = X.reshape(-1, 2)
xones = np.ones(X.shape[0]*2).reshape(X.shape[0],2) #.shape
X = np.concatenate((X, xones), axis=1)
X = X.reshape(-1,3,4)
```

```python
X.shape
```

```python
num_features = X.shape[2]

print('Number of features:', num_features)
```

## 3: Créer le model
___
time steps = 3 car on a 3 points de coordonnées en séquence

```python
hidden_units = 100
max_time_steps = 3

model = Sequential([
    SimpleRNN(hidden_units, input_shape=(None, num_features)),
    RepeatVector(max_time_steps),    
    SimpleRNN(hidden_units, return_sequences=False),    
    Dense(4, activation='linear') 
])

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.summary()
```

## 4: Training the Model


```python
simple_logger = LambdaCallback(
    on_epoch_end=lambda e, l: print('{}=>{:.2f}'.format(e, l['val_accuracy']), end=' _ ')
        
)
early_stopping = EarlyStopping(monitor='val_loss', patience=400)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y) 

epochs=2000

history = model.fit(X_train, y_train, epochs=epochs, validation_split=0.2, verbose=2,
        batch_size=100,
        #callbacks=[simple_logger, early_stopping]
        callbacks=[early_stopping]
)
```

```python
#history.history.keys()
```

#### Vérifier le résultat

```python
score, acc = model.evaluate(X_test, y_test)
print('Test score:', score)
print('Test accuracy:', acc)
```
